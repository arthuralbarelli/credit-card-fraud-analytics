```python
import pandas as pd

pd.set_option("display.max_columns", None)
```

```python
train_data_csv_path = "../data/fraudTrain.csv"
test_data_csv_path = "../data/fraudTest.csv"
```

```python
train_df = pd.read_csv(train_data_csv_path, index_col=0)
```

```python
test_df = pd.read_csv(test_data_csv_path, index_col=0)
```

```python
full_data_fraud = pd.concat([train_df, test_df])
```

```python
print(train_df.columns)
print(train_df.size)
```

```python
print(test_df.columns)
print(test_df.size)
```

```python
print(train_df['amt'].quantile([0.05, 0.25, 0.5, 0.75, 0.95]), end='\n\n' )
print(test_df['amt'].quantile([0.05, 0.25, 0.5, 0.75, 0.95]), end='\n\n')
print(full_data_fraud['amt'].quantile([0.05, 0.25, 0.5, 0.75, 0.95]))
```

We can notice that percentiles of 'amt' feature for train, test and full dataframes are almost the same. Showing that they have a similar distribution to this feature.
